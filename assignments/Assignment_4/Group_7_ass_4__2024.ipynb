```python
years = [2019, 2018, 2017, 2016, 2015]

all_data = {}

for year in years: 
    
    data_path = "../../_data/endes//2019/"
    
    rec_1, meta_1 = pyreadstat.read_sav(os.path.join(data_path, 'REC0111.sav')) 
    rec_2, meta_2 = pyreadstat.read_sav(os.path.join(data_path, 'RE223132.sav'))
    rec_3, meta_3 = pyreadstat.read_sav(os.path.join(data_path, 'RE516171.sav'))
   
    cols1 = ['CASEID', 'V000', 'V001', 'V002', 'V003', 'V004', 'V008', 'V009', 'V010', 'V011', 'V012', 'V024', 'V102', 'V120', 'V121', 'V122', 'V123', 'V124', 'V125', 'V127', 'V133']
       
    rec1_1 = rec_1.loc[:, cols1]
    
        
    cols2 = ['CASEID', 'V201', 'V218', 'V301', 'V302', 'V323', 'V323A', 'V325A', 'V326', 'V327', 'V337', 'V359', 'V360', 'V361', 'V362', 'V363', 'V364', 'V367', 'V372', 'V372A', 'V375A', 'V376', 'V376A', 'V379', 'V380']
    rec2_1 = rec_2.loc[:, cols2]

    cols3 = ['CASEID', 'V501', 'V502', 'V503', 'V504', 'V505', 'V506', 'V507', 'V508', 'V509', 'V510', 'V511', 'V512', 'V513', 'V525', 'V613', 'V714', 'V715']
    rec3_1 = rec_3.loc[:, cols3]
    
    
    rec1_1.loc[:, 'year'] = year
    
    var_labels = {} 
    var_labels.update(meta_1.column_names_to_labels)
    var_labels.update(meta_2.column_names_to_labels)
    var_labels.update(meta_3.column_names_to_labels)
    
    
    value_labels = {} 
    value_labels.update(meta_1.variable_value_labels)
    value_labels.update(meta_2.variable_value_labels)
    value_labels.update(meta_3.variable_value_labels)
    
    
    endes_data = rec1_1.merge(rec2_1, on='CASEID', how='inner')
    endes_data = endes_data.merge(rec3_1, on='CASEID', how='inner')
    
    all_data[f'year_{year}'] = {'data': endes_data, 'var_labels': var_labels, 'value_labels': value_labels}
all_data
```

```python

# Assuming 'all_data' dictionary has been created from the previous code

# Create an empty list to store individual DataFrames
all_dataframes = []

# Iterate over each year in 'all_data'
for year_key, year_data in all_data.items():
    # Extract the 'data' dictionary for the current year
    current_data = year_data["data"]

    # Iterate over each dataset in the 'data' dictionary
    for dataset_name, dataset_df in current_data.items():
        # Append the dataset DataFrame to the list
        all_dataframes.append(dataset_df)

# Concatenate all DataFrames in the list
endes_data_2015_2019 = pd.concat(all_dataframes, ignore_index=True)

# Display the resulting DataFrame
print(endes_data_2015_2019)

all_dataframes 
```

```python
# Assuming 'all_data' dictionary has been created from the previous code

# Create dictionaries to store variable labels and value labels
all_var_labels = {}
all_value_labels = {}

# Iterate over each year in 'all_data'
for year_key, year_data in all_data.items():
    # Extract the 'var_labels' and 'value_labels' dictionaries for the current year
    var_labels = year_data["var_labels"]
    value_labels = year_data["value_labels"]

    # Store the dictionaries in the main dictionaries
    all_var_labels[year_key] = var_labels
    all_value_labels[year_key] = value_labels

# Concatenate all DataFrames in the list
endes_data_2015_2019 = pd.concat(all_dataframes, ignore_index=True)

# Extract the year from the first key of 'all_var_labels'
year_key = list(all_var_labels.keys())[0]

# Add new attributes to 'endes_data_2015_2019' for var_labels and value_labels
endes_data_2015_2019['var_labels'] = all_var_labels[year_key]
endes_data_2015_2019['value_labels'] = all_value_labels[year_key]

# Display the resulting DataFrame with new attributes
print(endes_data_2015_2019)
```
