```python
# We select the specific columns
selected_columns = ['V201', 'V613', 'V715', 'V511']

# We get descriptive statistics and count missing values
result_df = endes_2019[selected_columns].describe().T
result_df['N_missing'] = endes_2019[selected_columns].isnull().sum()

# We rearrange the columns
result_df = result_df[['min', 'max', 'mean', 'count', 'N_missing']]

# Rename the columns
result_df.columns = ['Min', 'Max', 'Mean', 'N_obs', 'N_missing']

# We add the variable column
result_df['Variables'] = result_df.index

# We rearrange the columns
result_df = result_df[['Variables', 'Min', 'Max', 'Mean', 'N_obs', 'N_missing']]

# Sort by the number of missing values
result_df = result_df.sort_values(by='N_missing', ascending=False)

# Show the resulting DataFrame
print(result_df)
```
